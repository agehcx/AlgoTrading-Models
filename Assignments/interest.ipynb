```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
yr = 35
interest = 31

target = 20
```

```python
df = pd.DataFrame(dtype=np.float64)

df[1] = [1] * yr

for i in range (1,interest):
    df[i] = 1.00
    
df
```

```python
## Calculate Interest

for i in range (1,interest):
    for j in range (1,yr):
        df.loc[df.index[j], i] = df.loc[df.index[j - 1], i] * (1 + i / 100)
```

```python
pd.options.display.float_format = '{:.4f}'.format
df
```

```python
plt.figure(figsize=(20,6))
plt.plot(df[target], label=f"{target:.1f} annual return rate")
plt.xlabel("Time in years")
plt.ylabel("Returns")
plt.yscale('symlog')
plt.legend()
plt.show()
```
