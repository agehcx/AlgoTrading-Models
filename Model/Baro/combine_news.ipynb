```python
import pandas as pd
import numpy as np
```

```python
news_website = pd.read_csv("NEWS_WEBSITE_2024.csv")
```

```python
news_website.head()
```

```python
news_website = news_website.drop(columns=['NewsId'])
```

```python
news_website
```

```python
news_social = pd.read_csv("NEWS_SOCIAL_2024.csv")
```

```python
news_social.head()
```

```python
news_social = news_social.drop(columns=['FaceNewsId'])
```

```python
news_social
```

```python
news_combine = pd.concat([news_social, news_website], ignore_index=True)
```

```python
news_combine['CreateDate'] = pd.to_datetime(news_combine['CreateDate'], dayfirst=True, errors='coerce')

# Sort by CreateDate in ascending order
news = news_combine.sort_values(by='CreateDate', ascending=True)
```

```python
news['CreateDate'] = pd.to_datetime(news['CreateDate'])

# Create new 'Date' column
news['Date'] = news['CreateDate'].dt.date
```

```python
news
```

```python
def agg_stocks(group):
    stocks_set = set()
    for stock in group["Stock"]:
        stocks_set.add(stock)
    stocks_set = sorted(stocks_set)

    stocks_list = sorted(group["Stock"].tolist())
    total_count = len(stocks_list)
    max_mentioned = (
        max(set(stocks_list), key=stocks_list.count) if stocks_list else None
    )
    max_mentioned_count = stocks_list.count(max_mentioned)
    return pd.Series(
        [stocks_list, stocks_set, total_count, max_mentioned, max_mentioned_count],
        index=["StockList", "StockSet", "totalCount", "maxMentioned", "maxCount"],
    )
```

```python
grouped = news.groupby("Date").apply(agg_stocks).reset_index()

# Rename columns for clarity
grouped.rename(columns={"StockList": "stockList"}, inplace=True)
```

```python
grouped.head(10)
```

```python
# grouped.to_csv("counted_news.csv")
```

```python
shifted_df = grouped.shift(1)
shifted_df.iloc[-1] = grouped.iloc[-1]
shifted_df["Date"] = grouped["Date"]
```

```python
shifted_df.head(10)
```

```python
shifted_df["stockList"] = shifted_df["stockList"].apply(
    lambda x: x if x is not None else []
)
```

```python
shifted_df["3DaySet"] = [set() for _ in range(len(shifted_df))]
```

```python
for i in range(len(shifted_df)):
    p3set = set()
    for j in range(max(0, i - 2), i + 1):
        p3set.update(shifted_df.at[j, "stockList"])
    shifted_df.at[i, "3DaySet"] = sorted(p3set)
```
