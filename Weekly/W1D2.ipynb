```python
df['EMA-26'] = df['Close'].ewm(span = 26, adjust=False).mean()
```

```python
df
```

```python
df['EMA-12'] = df['Close'].ewm(span=12,adjust=False).mean()
```

```python
df
```

```python
df['MACD'] = df['EMA-12'] - df['EMA-26']
```

```python
df['Signal_Line'] = df['MACD'].ewm(span=9, adjust=False).mean()

df['Histogram'] = df['MACD'] - df['Signal_Line']
```

```python
df['Cross'] = 'No Cross'
```

```python
for i in range(1, len(df)):
    if df.iloc[i]['MACD'] > 0 and df.iloc[i-1]['MACD'] <= 0:
        df.at[df.index[i], 'Cross'] = 'MACD crossed above Signal'
    elif df.iloc[i]['MACD'] < 0 and df.iloc[i-1]['MACD'] >= 0:
        df.at[df.index[i], 'Cross'] = 'MACD crossed below Signal'
```

```python
plt.figure(figsize=(20, 7))
plt.plot(df.index, df['MACD'], label="MACD", color="green")
plt.plot(df.index, df['Signal_Line'], label="Signal", color="pink")
plt.title('MACD - SIGNAL LINE')
plt.xlabel('Date')
plt.ylabel('')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 7))
plt.plot(df.index, df['Close'], label='Close Price', color='black')
plt.plot(df.index, df['EMA-26'], label='EMA26', color='blue')
plt.plot(df.index, df['EMA-12'], label='EMA12', color='red')
# plt.plot(df.index, df['MACD'], label="MACD", color="green")
# plt.plot(df.index, df['Signal_Line'], label="Signal", color="pink")
# plt.plot(df.index, df['Histogram'], label="Histogram", color="purple")

# Highlight the cross points
cross_above = df[df['Cross'] == 'EMA26 crossed above EMA12']
cross_below = df[df['Cross'] == 'EMA26 crossed below EMA12']
plt.scatter(cross_above.index, cross_above['Close'], color='green', marker='^', alpha=1, label='EMA26 crossed above EMA12')
plt.scatter(cross_below.index, cross_below['Close'], color='red', marker='v', alpha=1, label='EMA26 crossed below EMA12')

plt.title('Close Price with EMA26 and EMA12')
plt.xlabel('Date')
plt.ylabel('')
plt.legend()
plt.grid(True)
plt.show()
```
